```python
x = sample_df.drop(columns=['label'])
y = sample_df['label']
x.shape,y.shape
```

```python
fig,axes = plt.subplots(2,3,figsize = (10,8))
axes = axes.flatten()
for i in range(5):
  mat = x.iloc[i]
  label = y.iloc[i]
  mat = mat.values
  mat = mat.reshape(28,28)
  axes[i].imshow(mat,cmap='gray')
  axes[i].set_title(f"Number: {label}")

plt.delaxes(axes[-1])
plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()

model.fit(x,y)
```

```python
y_pred = model.predict(train_df.drop(columns='label'))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(train_df['label'],y_pred)
```
