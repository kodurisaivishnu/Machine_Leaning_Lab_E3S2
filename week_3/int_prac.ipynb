```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tvmarketing.csv')

df.head()
```

```python
sns.regplot(x = df['TV'],y=df['Sales'],color='skyblue')
```

```python
b1 = np.cov(df['TV'],df['Sales'])[0][1]/np.var(df['TV'])
b0 = np.mean(df["Sales"] )  -  b1 * np.mean(df['TV'])

b0,b1
```

```python
num = np.sum((df['TV'] - df['TV'].mean()) * (df['Sales'] - df["Sales"].mean()))
den = np.sum((df['TV'] - df['TV'].mean())**2)
print("b1 :",num/den)
b0 = np.mean(df["Sales"] )  -  b1 * np.mean(df['TV'])
print("b0 :",b0)
```

```python
x = df['TV']
y = df['Sales']
x = x.values.reshape(-1,1)
x.shape
```
